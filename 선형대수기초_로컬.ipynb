```python
stu_height = np.array([166,168,170,172,174])
```

```python
# 표준편차 구하세요. (하나하나 계산해서, std와 비교 후 같도록)
```

```python
print(sum(stu_height)/len(stu_height))
```

```python
x_bar = np.mean(stu_height)
```

```python
x_bar   # 편차
```

```python
(stu_height-x_bar)**2
```

```python
sum((stu_height-x_bar)**2)
```

```python
sum((stu_height-x_bar)**2)/(len(stu_height)-1)   # 표본분산
```

```python
s_power = sum((stu_height-x_bar)**2)/(len(stu_height)-1)
```

```python
np.var(stu_height)
```

```python
np.sqrt(s_power)
```
